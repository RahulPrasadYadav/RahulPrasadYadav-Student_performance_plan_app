Classification

```python
df_encoded['pass'] = df_encoded['G3'].apply(lambda x: 1 if x >= 10 else 0)
```

```python
df_encoded.head()
```

```python
X=df_encoded.drop('pass',axis=1)
Y=df_encoded['pass']
```

```python
x_trainn,x_val,y_trainn,y_val=train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_model=RandomForestClassifier()
```

```python
clf_model.fit(x_trainn,y_trainn)
```

```python
y_pred_clg_reg=clf_model.predict(x_val)
```

```python
print("r2 score",r2_score(y_val,y_pred_clg_reg))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score


```

```python
print("Accuracy:", accuracy_score(y_val, y_pred_clg_reg))
print("\nConfusion Matrix:\n", confusion_matrix(y_val, y_pred_clg_reg))
print("\nClassification Report:\n", classification_report(y_val, y_pred_clg_reg))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf_model, X, Y, cv=5)
print("CV Accuracy Scores:", scores)
print("Average CV Accuracy:", scores.mean())
```

```python


feat_imp = pd.Series(clf_model.feature_importances_, index=X.columns)
feat_imp.nlargest(10).plot(kind='barh')
plt.title("Top 10 Important Features")
plt.show()
```

```python
def recommend_plan(row):
    plan = []
    if row['studytime'] <= 1:
        plan.append("📚 Increase study time.")
    if row['failures'] > 0:
        plan.append("🔁 Revise failed subjects.")
    if row['absences'] > 5:
        plan.append("📅 Attend more classes.")
    if row['goout'] > 3:
        plan.append("⚠️ Limit social outings.")
    if row['Dalc'] > 2 or row['Walc'] > 2:
        plan.append("🚫 Reduce alcohol consumption.")
    if not plan:
        plan.append("✅ Keep up the good work!")
    return " | ".join(plan)

df_encoded['learning_plan'] = df_encoded.apply(recommend_plan, axis=1)
```

```python
df_encoded[['G3', 'pass', 'learning_plan']].to_csv('student_learning_output.csv', index=False)
```
